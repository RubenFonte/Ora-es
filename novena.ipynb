# Novena à Imaculada Conceição

## Estrutura diária
- **Sinal da Cruz**
- **Oração inicial**
- **Meditação do dia**
- **Pai-Nosso, Ave-Maria, Glória**
- **Oração final**

---

## Dia 1
**Oração inicial**  
Ó Maria Imaculada, escolhida por Deus desde toda a eternidade, venho a Ti com confiança.  
Intercede por mim junto ao Teu Filho Jesus, para que eu seja purificado e fortalecido na fé.  

**Meditação**  
Maria foi preservada do pecado original.  
Peço a graça de viver com coração puro e aberto à vontade de Deus.  

**Pai-Nosso, Ave-Maria, Glória**  

**Oração final**  
Imaculada Conceição, rogai por nós.
