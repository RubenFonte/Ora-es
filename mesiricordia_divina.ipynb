# 🙏 Novena da Divina Misericórdia
*Pedida por Jesus a Santa Faustina Kowalska*

A novena inicia-se tradicionalmente na **Sexta-Feira Santa**, preparando o fiel para a **Festa da Misericórdia** (celebrada no domingo seguinte ao de Páscoa). No entanto, ela pode ser realizada em qualquer período de necessidade espiritual.

---

## 📅 Roteiro das Intenções Diárias
Jesus pediu que, a cada dia, mergulhássemos um grupo de almas em Sua Misericórdia:

| Dia | Intenção Principal |
| :--- | :--- |
| **1º Dia** | Toda a humanidade, especialmente os pecadores. |
| **2º Dia** | As almas dos sacerdotes e religiosos. |
| **3º Dia** | As almas devotas e piedosas. |
| **4º Dia** | Os pagãos e aqueles que ainda não conhecem a Jesus. |
| **5º Dia** | As almas dos cristãos separados da Igreja. |
| **6º Dia** | As almas mansas, humildes e as das criancinhas. |
| **7º Dia** | As almas que veneram e glorificam a Minha misericórdia. |
| **8º Dia** | As almas que estão no Purgatório. |
| **9º Dia** | As almas tíbias (frias na fé). |

---

## 📿 Como Rezar o Terço da Misericórdia
*O Terço é a oração central que deve ser repetida após a intenção de cada dia.*

1.  **Orações Iniciais:**
    * Pai Nosso
    * Ave Maria
    * Credo (Creio em Deus Pai...)
2.  **Nas contas do "Pai Nosso" (Contas Grandes):**
    > "Eterno Pai, eu Vos ofereço o Corpo e o Sangue, a Alma e a Divindade de Vosso diletíssimo Filho, Nosso Senhor Jesus Cristo, em expiação dos nossos pecados e dos do mundo inteiro."
3.  **Nas contas da "Ave Maria" (Contas Pequenas):**
    > "Pela Sua dolorosa Paixão, tende misericórdia de nós e do mundo inteiro."
4.  **Ao final (Oração Final - Repetir 3 vezes):**
    > "Deus Santo, Deus Forte, Deus Imortal, tende piedade de nós e do mundo inteiro."
5.  **Fechamento (Opcional):**
    > "Ó Sangue e Água que jorrastes do Coração de Jesus como fonte de misericórdia para nós, eu confio em Vós!"

---

## 💡 A "Hora da Misericórdia"
Jesus destacou a importância das **15:00h** (a hora de Sua morte na cruz). Ele prometeu que nada seria negado à alma que pedisse em nome de Sua Paixão nesse horário, desde que o pedido fosse conforme a vontade divina.

> **Dica:** Se você estiver no trabalho ou ocupado às 15h, basta uma breve elevação do pensamento: *"Jesus, eu confio em Vós"*.
